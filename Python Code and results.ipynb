```python
import pandas as pd

df = pd.read_csv("patients.csv")

df.head()
```

```python
print("Disease Count:")
print(df['disease'].value_counts())

print("\nAverage Treatment Cost:")
print(df['treatment_cost'].mean())

print("\nDoctor Workload:")
print(df['doctor'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Pie Chart
df['disease'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Disease Distribution")
plt.ylabel('')
plt.show()

# Bar Chart
df['doctor'].value_counts().plot(kind='bar')
plt.title("Doctor Workload")
plt.xlabel("Doctor")
plt.ylabel("Patients")
plt.show()
```
